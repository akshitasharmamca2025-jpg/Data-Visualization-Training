```python

#41.student exam scores:
# Import matplotlib library
import matplotlib.pyplot as plt

# Student exam scores data
scores = [45, 50, 55, 60, 62, 65, 68, 70, 72, 75, 78, 80, 83, 85, 90]

# Create histogram
# bins=5 means data will be divided into 5 intervals (groups)
plt.hist(scores, bins=5)

# Add title
plt.title("Student Exam Scores Distribution")

# Label X-axis
plt.xlabel("Scores")

# Label Y-axis
plt.ylabel("Number of Students")

# Add grid for better understanding
plt.grid()

# Display the histogram
plt.show()
```
